```python
import pandas as pd

data = {
    "CustomerID": [1001,1002,1003,1004,1005,1006,1007,1008,1009,1010],
    "Age": [25,34,45,23,36,52,29,41,31,27],
    "Tenure_Months": [12,24,6,3,18,36,8,30,15,5],
    "Monthly_Charges": [50,70,90,40,60,80,55,75,65,45],
    "Support_Tickets": [1,0,5,4,2,0,3,1,2,6],
    "Login_Frequency": [25,30,8,5,20,28,10,24,18,3],
    "Subscription_Type": ["Basic","Premium","Premium","Basic","Standard","Premium","Basic","Premium","Standard","Basic"],
    "Churn": ["No","No","Yes","Yes","No","No","Yes","No","No","Yes"]
}

df = pd.DataFrame(data)
df.to_csv("customer_churn.csv", index=False)

print(df)
```

```python
import pandas as pd

df = pd.read_csv("customer_churn.csv")

print(df.head())
```

```python
print(df.info())
```

```python
print(df.isnull().sum())
```

```python
print(df["Churn"].value_counts())
```

```python
print(df.groupby("Churn")[["Tenure_Months",
                           "Monthly_Charges",
                           "Support_Tickets",
                           "Login_Frequency"]].mean())
```

```python
print(pd.crosstab(df["Subscription_Type"], df["Churn"]))
```

```python
import matplotlib.pyplot as plt

df["Churn"].value_counts().plot(kind="bar")
plt.title("Customer Churn Distribution")
plt.xlabel("Churn")
plt.ylabel("Count")
plt.show()
```
