```python
import pandas as pd
import folium
from folium.plugins import HeatMap
import plotly.express as px
```

```python
data = {
    'Region': ['Hyderabad', 'Bangalore', 'Chennai', 'Mumbai', 'Pune', 'Delhi'],
    'Latitude': [17.3850, 12.9716, 13.0827, 19.0760, 18.5204, 28.7041],
    'Longitude': [78.4867, 77.5946, 80.2707, 72.8777, 73.8567, 77.1025],
    'Sales': [50000, 70000, 40000, 85000, 30000, 90000],
    'Stores': [5, 8, 2, 10, 1, 12]
}

df = pd.DataFrame(data)

print(df)
```

```python
df['Opportunity_Score'] = df['Sales'] / (df['Stores'] + 1)

print("Best Areas for Expansion:")
print(df.sort_values(by='Opportunity_Score', ascending=False))
```

```python
m = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

for _, row in df.iterrows():
    folium.Marker(
        [row['Latitude'], row['Longitude']],
        popup=f"{row['Region']}<br>Sales: {row['Sales']}<br>Stores: {row['Stores']}",
        tooltip=row['Region']
    ).add_to(m)

m
```

```python
heat_data = df[['Latitude', 'Longitude', 'Sales']].values.tolist()

heatmap = folium.Map(location=[20.5937, 78.9629], zoom_start=5)

HeatMap(heat_data).add_to(heatmap)

heatmap
```

```python
fig = px.bar(
    df.sort_values('Opportunity_Score', ascending=False),
    x='Region',
    y='Opportunity_Score',
    title='Business Expansion Opportunities'
)

fig.show()
```

```python
m.save("business_expansion_map.html")
print("File saved as business_expansion_map.html")
```
